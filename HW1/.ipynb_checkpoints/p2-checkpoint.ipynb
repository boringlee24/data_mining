```python
%matplotlib inline
import numpy as np
import pdb
import sklearn.metrics.pairwise as distance
import matplotlib
#matplotlib.use('tkAgg')
import matplotlib.pyplot as plt
import glob
import sys
from matplotlib.ticker import MultipleLocator
import math
```

```python
#%%script false --no-raise-error
######## part (a) #############
fig, axs = plt.subplots(1, 1, gridspec_kw={'hspace': 0.1, 'wspace': 0.2, 'bottom': 0.13, 
                                           'top': 0.9, 'right':0.995, 'left':0.17}, figsize=(5,3))
axs.set_title(f'Euclidean Distance', fontsize=14)
n_list = [100, 1000]
for n in n_list:
    data_x = np.arange(100) + 1
    data_y = []
    index = n_list.index(n)
    for k in range(1, 101): # dimensionality
        X = [] # len=n, each has k dimension
        for i in range(n):
            X.append(np.random.uniform(size=k))
        euclid = distance.euclidean_distances(X)
        euclid = euclid.flatten()
        non_zero = [k for k in euclid if k!=0]
        rk = (max(non_zero) - min(non_zero)) / np.mean(non_zero)
        rk = math.log10(rk)
        data_y.append(rk)
    axs.plot(data_x, data_y, label=f'n = {n}')
axs.set_ylabel('R(k) value', fontsize=14)
axs.set_xlabel('k value', fontsize=14)
axs.grid(which='major', axis='both', ls='dashed')
axs.legend(loc='upper right', fontsize=14)
fig.savefig(f'plots/p2_euclid.png', bbox_inches='tight')
    
```

```python
######## part (b) #############
fig, axs = plt.subplots(1, 1, gridspec_kw={'hspace': 0.1, 'wspace': 0.2, 'bottom': 0.13, 
                                           'top': 0.9, 'right':0.995, 'left':0.17}, figsize=(5,3))
axs.set_title(f'Cosine Distance', fontsize=14)
n_list = [100, 1000]
for n in n_list:
    data_x = np.arange(99) + 2
    data_y = []
    index = n_list.index(n)
    for k in range(2, 101): # dimensionality
        X = [] # len=n, each has k dimension
        for i in range(n):
            X.append(np.random.uniform(size=k))
        cosine = distance.cosine_distances(X)
        cosine = cosine.flatten()
        non_zero = [k for k in cosine if k!=0] 
        rk = (max(non_zero) - min(non_zero)) / np.mean(non_zero)
        rk = math.log10(rk)
        data_y.append(rk)
    axs.plot(data_x, data_y, label=f'n = {n}')
axs.set_ylabel('R(k) value', fontsize=14)
axs.set_xlabel('k value', fontsize=14)
axs.grid(which='major', axis='both', ls='dashed')
axs.legend(loc='upper right', fontsize=14)
fig.savefig(f'plots/p2_cosine.png', bbox_inches='tight')
```

```python
######## part (c) #############
fig, axs = plt.subplots(1, 1, gridspec_kw={'hspace': 0.1, 'wspace': 0.2, 'bottom': 0.13, 
                                           'top': 0.9, 'right':0.995, 'left':0.17}, figsize=(5,3))
axs.set_title(f'Euclidean Distance', fontsize=14)
n_list = [100, 1000]
mu, sigma = 0, 1
for n in n_list:
    data_x = np.arange(100) + 1
    data_y = []
    index = n_list.index(n)
    for k in range(1, 101): # dimensionality
        X = [] # len=n, each has k dimension
        for i in range(n):
            X.append(np.random.normal(mu, sigma, size=k))
        euclid = distance.euclidean_distances(X)
        euclid = euclid.flatten()
        non_zero = [k for k in euclid if k!=0]
        rk = (max(non_zero) - min(non_zero)) / np.mean(non_zero)
        rk = math.log10(rk)
        data_y.append(rk)
    axs.plot(data_x, data_y, label=f'n = {n}')
axs.set_ylabel('R(k) value', fontsize=14)
axs.set_xlabel('k value', fontsize=14)
axs.grid(which='major', axis='both', ls='dashed')
axs.legend(loc='upper right', fontsize=14)
fig.savefig(f'plots/p2c_euclid.png', bbox_inches='tight')
```
